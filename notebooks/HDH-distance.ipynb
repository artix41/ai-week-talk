```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

from sklearn.datasets import make_blobs
```

```python
colormap = np.array(plt.rcParams['axes.prop_cycle'].by_key()['color'])
```

```python
N_points = 50
X_source_0 = make_blobs(N_points, centers=2*np.array([[-1,-1]]), cluster_std=0.4)[0]
X_source_1 = make_blobs(N_points, centers=2*np.array([[-1,1]]), cluster_std=0.4)[0]

X_target_0 = make_blobs(N_points, centers=2*np.array([[1,-1]]), cluster_std=0.4)[0]
X_target_1 = make_blobs(N_points, centers=2*np.array([[1,1]]), cluster_std=0.4)[0]
```

```python
plt.rcParams['figure.figsize'] = (12, 10)
plt.axis("equal")
plt.axis([-4,4,-4,4])

plt.scatter(X_source_0[:,0], X_source_0[:,1], color=colormap[1], s=70, label="Source")
plt.scatter(X_source_1[:,0], X_source_1[:,1], color=colormap[0], s=70)

plt.scatter(X_target_0[:,0], X_target_0[:,1], marker="x", s=70, color=colormap[0], label="Target")
plt.scatter(X_target_1[:,0], X_target_1[:,1], marker="x", s=70, color=colormap[0])

plt.plot([-10, 10], [8, -12], "--", color="black")

plt.legend(prop={'size': 15})

ax = plt.gca()
leg = ax.get_legend()
leg.legendHandles[0].set_color('black')
leg.legendHandles[1].set_color('black')
```
